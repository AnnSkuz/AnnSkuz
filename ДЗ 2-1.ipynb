# Задание 1

Необходимо вывести названия треков, которые были приобретены компаниями, которые обслуживаются сотрудниками, нанятыми после 1 сентября 2002 года. 
    

```python
%%sql
select t.name, invoiceDate, company
from tracks t, invoice_items inv_i, invoices inv, customers c
where t.trackid=inv_i.trackid
and inv_i.invoiceid=inv.invoiceid
and inv.customerid=c.customerid
and  invoiceDate>'2002-09-01' 
```

## Задание 2
Вывести общую стоимость и продолжительность каждого плейлиста.

```python
%%sql
select p.name, sum(unitprice) as price, sum(milliseconds) as time
from tracks t, playlist_track p_t, playlists p
where t.trackid=p_t.trackid
and p_t.playlistid=p.playlistid
group by p_t.playlistid
```

## Задание 3
Для каждого артиста определить все жанры песен, которые у него есть и количество песен по каждому жанру.


```python
%%sql
select a.name as artists_name, g.name as genres_name, count(t.name) as tracks 
from artists a, albums al, tracks t, genres g
where a.artistid=al.artistid
and al.albumid=t.albumid
and t.genreid=g.genreid
group by a.name, g.name 
```

## Задание 4
Для каждого артиста найти жанр, по которому он заработал наибольшее количество денег, а также вывести соотношение общей суммы, заработанной артистом по всем другим жанрам к общей сумме заработанных денег.

```python
%%sql 
select a.name, g.name as genre, sum(t.unitprice) as price
from artists a, albums al, tracks t, genres g, invoice_items inv_i
where a.artistid=al.artistid
and al.albumid=t.albumid
and t.genreid=g.genreid
and t.trackid=inv_i.trackid
group by a.name, g.name
```

## Задание 5
Для каждой компании вывести общее количестов песен, купленных по жанру рок

```python
%%sql 
select company, count(t.name) as tracks
from tracks t, genres g, invoice_items inv_i, invoices inv, customers c
where t.trackid=inv_i.trackid
and inv_i.invoiceid=inv.invoiceid
and t.genreid=g.genreid
and inv.customerid=c.customerid
and g.name='Rock'
group by company
```
